```python
numerical_col= ["BudgetUSD","US_BoxOfficeUSD","Global_BoxOfficeUSD","Opening_Day_SalesUSD","One_Week_SalesUSD","IMDbRating","RottenTomatoesScore","NumVotesIMDb","NumVotesRT"]
```

```python
for col in numerical_col:
    print(col)
```

```python
for col in numerical_col:
    sns.histplot(mv[col],bins=50, kde=True)
    plt.title(f"distribution of {col}")
    plt.show()
```

```python
cate_col=["Genre","Country","Director","LeadActor"]
```

```python
for i in cate_col:
    mv[i].value_counts().head(20).plot(kind="bar",)
    plt.title(f"top 20 {i} counts")
    plt.show()
```

```python
plt.figure(figsize=(6,3),facecolor="lightgreen")
sns.scatterplot(data=mv, x="BudgetUSD", y="Global_BoxOfficeUSD", alpha=0.3)
plt.title("Budget vs Global counts")
plt.show()
```

```python
mv.columns
```

```python
genra_year_stats=mv.groupby(["Genre","ReleaseYear"]).agg({"Global_BoxOfficeUSD":"mean","BudgetUSD":"mean","IMDbRating":"mean"})
```

```python
genra_year_stats
```

```python
plt.figure(figsize=(8,4),facecolor="lightblue")
sns.lineplot(data=genra_year_stats, x="ReleaseYear", y="Global_BoxOfficeUSD",hue="Genre",palette="rainbow")
plt.title("Release year vv global box office usd")
plt.show()
```

```python
plt.figure(figsize=(8,4),facecolor="lightblue")
sns.lineplot(data=genra_year_stats.query("Genre =='Action'"), x="ReleaseYear", y="Global_BoxOfficeUSD",hue="Genre",color="blue")
plt.title("Release year vv global box office usd")
plt.show()
```

```python
top_actors= mv.groupby("LeadActor")["IMDbRating"].mean().sort_values(ascending=False).head(20)
```

```python
top_actors
```

```python
plt.figure(figsize=(8,4),facecolor="skyblue")
top_actors.plot(kind="bar",color="lightgreen")
plt.title("top avtors")
plt.show()
```
