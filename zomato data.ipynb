```python
country=pd.read_excel("Country-Code.xlsx")
```

```python
country.head()
```

```python
final_df=pd.merge(df,country,on='Country Code', how='left') # zomato data set and country code data set ko combine kiya hai ek data ftrame me 
```

```python
final_df.head()
```

```python
country_name=final_df.Country.value_counts().index  # country ke names hai 
```

```python
country_value=final_df.Country.value_counts().values   # country ki values hai isle base me transactions nikalne hai okk
```

```python
plt.pie(country_value[:3],labels=country_name[:3],autopct="%1.2f%%") # top 3 countrys jaha zomato jyada use hota hai for food orderd 
```

OBSERVATION -- ZOMATOS MAXIMUM TRASACTIONS FROM INDIA THATS AROUND 94.39 PERCENT,, SECOND IS UNITED STATE THATS AROUND 4.73 PERCENT
LAST IS UNITED KINGDOM THATS AROUND 0.87 PERCENT
