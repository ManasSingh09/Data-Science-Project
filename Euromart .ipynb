```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv("EuroMart Stores.csv")
```

```python
df.groupby("Country")[" Sales "].mean()
```

```python
sns.barplot(data=df, x="Country", y=" Sales ",hue="Country", palette="rainbow")
plt.title("Country wise Sales")
plt.xlabel("Country")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
top_10_city=df.groupby("City")[" Sales "].mean().head().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,4),facecolor="lightgreen")
df.groupby("City")[" Sales "].sum().head(10).plot(kind="bar")
plt.title("top 10 cities sales")
plt.xticks(rotation=45)
plt.ylabel("Sales")
plt.show()
```

```python
df.groupby("Category")[" Sales "].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,4),facecolor="lightyellow")
df.groupby("Category")[" Sales "].mean().plot(kind="pie",autopct="%1.2f%%")
plt.title("Category wise Sales")
plt.ylabel(" ")
plt.show()
```

```python
df.groupby("Region")[" Sales "].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(4,3),facecolor="lightgreen")
sns.barplot(data=df, x="Region", y=" Sales ")
plt.title("Region wise Sales")
plt.show()
```
