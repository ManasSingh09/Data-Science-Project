```python
wd.groupby("country")["world percentage"].sum()
```

```python
sns.histplot(wd["2023 population"], bins =10, kde=True)
plt.title("distribution od 2023 population")
plt.xlabel("frequency")
plt.ylabel("population")
plt.show()
```

```python
wd.groupby("continent")["2023 population"].mean().sort_values(ascending=False)
```

```python
top_10_country=wd.sort_values(by="2023 population",ascending = False).head(10)
```

```python
plt.figure(figsize=(4,3),facecolor="yellow")
sns.barplot(data=top_10_country, x="2023 population", y="country")
plt.title("top 10 countryes population")
plt.xlabel("population")
plt.ylabel("country")
plt.show()
```

```python
wd.columns
```

```python
sns.scatterplot(x="2023 population", y="descity(km2)", data=wd)
plt.title("poppulation and density")
plt.xlabel("density")
plt.ylabel("population")
plt.show()
```

```python
wd.groupby("continent")["area (km2)"].sum().sort_values(ascending=False)
```

```python
sns.violinplot(x="continent", y="descity(km2)", data=wd, palette = "viridis")
plt.title("population density distribution asross continent")
plt.xlabel("continent")
plt.ylabel("population of density")
plt.show()
```

```python
sns.boxplot(x="continent", y="descity(km2)", data=wd, palette="muted",showfliers=False)
plt.title("box plot of descity distribution across continent (withought outliers)")
plt.show()
```
